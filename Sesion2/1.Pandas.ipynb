```python
data1 = {
    'A':['A0','A1','A3'],
    'KEY':['K0','K1','K2']
}

data2 = {
    'B':['B0','B1','B3'],
    'KEY':['K3','K1','K2']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)


df1.merge(df2, on='KEY', how='left')
```

```python
pd.merge(df1,df2, on ='KEY')
```

```python
##CUANDO LAS CLAVES SON DIFERENTES


data1 = {
    'A':['A0','A1','A3'],
    'KEY1':['K0','K1','K2']
}

data2 = {
    'B':['B0','B1','B3'],
    'KEY2':['K3','K1','K2']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)


df1.merge(df2, left_on='KEY1', right_on='KEY2')
```
